```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression # THIS IS FOR THE MATHCNNE LEARNING MODEL
from sklearn.model_selection import train_test_split # tHIS IS FOR TEH SPLITING THE MODEL IN OT THE SPEARET FILE 
from sklearn.metrics import mean_squared_error , r2_score # To measure how good our model is
import matplotlib.pyplot as plt
```

```python
house_distance = np.array([500 , 600 , 400 , 200 , 400 ]).reshape(-1 , 1)
house_price = np.array([150 , 200 , 450 , 65 ,85])

print("The size of house is : " , house_distance.flatten())
print("The hosue price: " , house_price)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(house_distance , house_price , color ='blue', alpha=0.7 , s=100)
plt.title("House price vs actual size")
plt.xlabel('house_distance')
plt.ylabel('house_price')
plt.grid(True , alpha = 0.7)
plt.show()
```
